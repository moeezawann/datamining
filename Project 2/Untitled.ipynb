```python
%pip install requests bs4 tqdm
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.basketball-reference.com/contracts/players.html'

page = requests.get(url)
page.encoding = 'utf-8'  # Explicitly setting encoding rule because names like "Nikola Jokić" were printing out like "Nikola JokiÄ" 


soup = BeautifulSoup(page.text, 'html')

```

```python
print(soup)
```

```python
soup.find('table', class_ = 'sortable stats_table' ) 
```

```python
rows = soup.find_all('tr')
players_name = []
# Loop through each row and find the player name
for row in rows:
    player_cell = row.find('td', {'data-stat': 'player'}) 
    if player_cell:
        players_name = player_cell.get_text()
        
name = "Tyrese Maxey"
if name in players_name:
    print("yes")
        
```
